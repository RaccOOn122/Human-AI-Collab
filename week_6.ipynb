```python
# Import libraries
import wfdb
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
# Load example ECG record and annotation from MIT-BIH
# Record 100 is a standard example with annotations
record = wfdb.rdrecord('100', pn_dir='mitdb')
annotation = wfdb.rdann('100', 'atr', pn_dir='mitdb')

# Display basic info
print('Signal shape:', record.p_signal.shape)
print('Sampling frequency:', record.fs)
print('Signal names:', record.sig_name)
```

```python
# Plot ECG signal with annotations
plt.figure(figsize=(15, 4))
plt.plot(record.p_signal[:3000, 0], label='ECG Lead 1')
for idx, sample in enumerate(annotation.sample):
    if sample < 3000:
        plt.axvline(x=sample, color='r', linestyle='--', alpha=0.5)
plt.title('ECG Segment with Beat Annotations')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude (mV)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Display first 10 beat annotations
for i in range(10):
    print(f"Sample: {annotation.sample[i]}, Symbol: {annotation.symbol[i]}")
```
